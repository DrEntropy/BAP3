## Exercise 7: 
Following the example in Figure 1.9, use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes, like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. What do you observe?

The variables here are $\mu$, $\sigma$, and $\alpha$.  Alpha is the 'skew'. 


```python
skew1= pz.SkewNormal(0,1,1)
skew1.plot_pdf(moments=["m", "d", "s", "k"])
```

```python
def recover_moments(dist,samples):
    samples = dist.rvs(samples)
    return (np.mean(samples), np.std(samples))

for n in [10,100,1000,10000]:
    print(f"recovered: {recover_moments(skew1,n)}")
 
```

Lets try an even more skewed examples

```python
skew2= pz.SkewNormal(0,1,10)
skew2.plot_pdf(moments=["m", "d", "s", "k"])
```
